# LINEAR REGRESSION 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import seaborn as sns 
```

```python
BostonTrain = pd.read_csv("boston_train.csv") 
BostonTrain=BostonTrain.drop('ID', axis = 1)
y = BostonTrain.medv 
X = BostonTrain.drop('medv', axis = 1)
```

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3) 
```

```python
model = LinearRegression()
model.fit(X_train,y_train) 
```

```python
predictions = model.predict(X_test)
from sklearn import metrics
MSE=metrics.mean_squared_error(y_test,predictions)
print("MSE",MSE)
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.show() 
```

```python
print('MAE:', 
metrics.mean_absolute_error(y_test,predictions))
print('MSE:',
metrics.mean_squared_error(y_test,predictions))
print('RMSE:',
np.sqrt(metrics.mean_squared_error(y_test,predictions))) 
# distribution of the target variable MEDV. We will use the distplot function from the seaborn library.
sns.distplot(y, bins=30)
plt.show()
```
