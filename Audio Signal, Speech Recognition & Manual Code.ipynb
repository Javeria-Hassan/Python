# Audio Signal

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import wavfile
freq,signal=wavfile.read("file_example_WAV_1MG.wav")
print("Signal Data Type: ",signal.dtype)
print("Signal Duration: ",round(signal.shape[0]/float(freq),2),'seconds')
signal=signal/np.power(2,15)
signal=signal[:100]
time=1000*np.arange(0,len(signal),1)/float(freq)
plt.plot(time,signal,color="blue")
plt.xlabel('time in MilliSeconds')
plt.ylabel('Amplitutde')
plt.title("Audio Signal Graph")
plt.show()
```

# Generate Your own Audio Signal 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
output="finallab.wav"
duration=5
freq_samp=5000
freq_tone=1500
min=-4*np.pi
max=4*np.pi
t=np.linspace(min,max,duration*freq_samp)
audio_signal=np.cos(2*np.pi*freq_tone*t)
write(output,freq_samp,audio_signal)
audio_signal=audio_signal[75:150]
time_axis=1000*np.arange(0,len(audio_signal),1)/float(freq_samp)
plt.plot(time_axis,audio_signal,color='red')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('generated audio signal')
plt.show()
```
